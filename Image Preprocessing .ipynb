```python
import cv2
```

```python
import numpy as np
```

Custom Array

```python
white = np.full((400,400),255,dtype = np.uint8)
Black = np.full((400,400),0,dtype = np.uint8)
gray = np.full((400,400),113,dtype = np.uint8)
```

```python
cv2.imshow("White",white)
cv2.imshow("Black",Black)
cv2.imshow("Gray",gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

Resize

```python
img_robot = cv2.imread("Robot.png")
img_robot.shape
```

```python
resize_img_robot = cv2.resize(img_robot,(500,600))
```

```python
resize_img_robot.shape
```

```python
resize_img_robot
```

```python
cv2.imshow("Robot",resize_img_robot)
cv2.waitKey()
cv2.destroyAllWindows()
```

BGR COLORS & Split

```python
b,g,r = cv2.split(resize_img_robot)
```

```python
b.shape
```

```python
z = np.full((600,500),0,dtype = np.uint8)
```

```python
z
```

```python
blue = cv2.merge([b,z,z])
```

```python
cv2.imshow("Blue",blue)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
blueness = cv2.merge([b,z,z])
greenness = cv2.merge([z,g,z])
redness = cv2.merge([z,z,r])

cv2.imshow("Blueness",blueness)
cv2.imshow("greenness",greenness)
cv2.imshow("redness",redness)
cv2.waitKey()
cv2.destroyAllWindows()
```

Converting BGR to RGB

```python
rgb = cv2.cvtColor(resize_img_robot,cv2.COLOR_BGR2RGB)
```

```python
rgb.shape
```

```python
cv2.imshow("rgb",rgb)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
r,g,b = cv2.split(rgb)
```

```python
red = cv2.merge([z,z,r])
cv2.imshow("redness",red)
cv2.waitKey()
cv2.destroyAllWindows()
```

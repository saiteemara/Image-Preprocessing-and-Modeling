##  Separate Train set and Test set

✅ What is StandardScaler?

StandardScaler is a preprocessing tool in scikit-learn that standardizes numerical features.
*So after scaling:

* ✅ mean becomes 0
* ✅ standard deviation becomes 1

✅ What is fit_transform()?

* Think of scaling like making all numbers similar size.

* fit_transform() does two steps at the same time:

✅ Step 1 — fit

* It learns from the training data:

* what is the average value (mean)

* how spread out the values are (standard deviation)

✅ Step 2 — transform

* It changes the data using those learned values.

* So the numbers become:

* mean = 0

* standard deviation = 1

fit()
* learns the statistics(mean & std) from the data. --> Only on training data
  
transform()
*  uses the learned values (mean & std) to scale the data. --> On train, test, new data
  
* fit_transform() --> fit + transform together --> Only for training data

```python
sc = StandardScaler()
```

```python
X_train_transformed = sc.fit_transform(X_train)
```

```python
X_test_transformed = sc.transform(X_test)
```

## Train a model

```python
RF = RandomForestClassifier(n_estimators=300)
```

```python
Model = RF.fit(X_train_transformed,y_train)
```

```python
y_pred = Model.predict(X_test_transformed)
```

```python
print("RandomForest = ",accuracy_score(y_test,y_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn_model = knn.fit(X_train_transformed,y_train)
```

```python
y_pred_knn = knn_model.predict(X_test_transformed)
```

```python
print("KNN = ",accuracy_score(y_test,y_pred_knn))
```
